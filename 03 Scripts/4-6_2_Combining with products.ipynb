```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import datasets
path = r'C:\Users\raque\Documents\08-2023 Instacart Basket Analysis'
df_ords_prods_comb = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

#### Combining orders_products_combined with products_checked

```python
# Merging the datasets 
df_merged = df_prods.merge(df_ords_prods_comb, on = 'product_id', indicator = 'merge2')
```

```python
df_merged.head()
```

```python
df_merged.shape
```

```python
# value_counts sums up all the value in the "merge2" column to see if it's a full match or not
df_merged['merge2'].value_counts()
```

#### Checking frequency of merge using an outer merge

```python
# Merging the datasets with outer 
df_merged_outer = df_prods.merge(df_ords_prods_comb, on = 'product_id', indicator = 'merge2', how = 'outer')
```

```python
df_merged_outer.head()
```

```python
# value_counts sums up all the value in the "merge2" column to see if it's a full match or not
df_merged_outer['merge2'].value_counts()
```

```python
df_merged_outer.shape
```

#### Exporting dataset

```python
# Export data to pkl
df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```

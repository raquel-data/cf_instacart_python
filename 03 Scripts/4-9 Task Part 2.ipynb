### Sampling data & line chart

```python
len(big) + len(small)
```

### Histogram "order_hour_of_the_day"

### Line chart to analyze expenditure depending on the hour of the day

##### Creating a data sample

```python
np.random.seed(4)
```

```python
# Storing a list of boolean in dev to assign which columns stay in the sample and which ones go
dev2 = np.random.rand(len(ords_prods_all)) <= 0.7
```

```python
# checking that big and small add up to the whole dataframe
len(ords_prods_all)
```

### Line chart to analyze the connection between age and no of dependents

```python
# Create dataframe with only the columns needed for the line chart to optimize the performance and speed
ords_prods_all_sample6 = small2[['age', 'n_dependants']]
```

```python
# Create dataframe grouping by
ords_prods_all_sample6_2 = ords_prods_all_sample6.groupby('age').sum()['n_dependants'].reset_index()
```

There isn't a clear correlation between age and number of dependants. 

### Scatterplot exploring connection between age and income

```python
# Creating a scatterplot
scatter = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_all)
```

In the scatterplot we observe that most people is on lower incomes than 200000. And getting the most scattered over 300000. We can see how +40 years old are the only ones to have an income over 400000 and way more incomes over 200000, than people in their 20's and 30's.
